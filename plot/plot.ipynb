```python
import math
import numpy as np
import pickle
def normalization1(data):
    data = data - np.min(data)
    data = data/(np.max(np.abs(data), axis=1)[:,np.newaxis]+0.00001)
    return data
def c_mass(data):
    interval = 1
    mass_max = 0
    mass_loc = 0
    for i in range(math.ceil(len(data)-interval/2)):
        if np.sum(data[i:i+interval]) >= mass_max:
            mass_max = np.sum(data[i:i+interval])
            mass_loc = i + interval/2
    return (mass_loc, mass_max)
def sorting(base):
    B = np.argsort(base.sum(1))
    B = B[::-1]
    return B


with open('/home/jiashuncheng/code/MANN/plot/data/rsc_h_norm_witchout_b.pkl', 'rb') as a:
    x = pickle.load(a)
Y = np.stack(x).squeeze(2)[:,0,:].T
Y = normalization1(Y)

fig, ax = plt.subplots()
y = Y[:,-15:]
B= sorting(y)
y = y[B,:]
print(y.shape)
plt.imshow(y, aspect='auto', origin='lower', vmin=0, vmax=1)
plt.colorbar()
plt.show()

fig, ax = plt.subplots()
y = Y[:,:15]
y = y[B,:]
plt.imshow(y, aspect='auto', origin='lower', vmin=0, vmax=1)
plt.colorbar()
plt.show()
```
