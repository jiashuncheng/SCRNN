```python
import pickle
import numpy as np
from sklearn.decomposition import PCA
from sklearn.manifold import Isomap
import matplotlib.pyplot as plt
from dPCA import dPCA
import math

class Bezier:
    # Pointsarray,num
    def __init__(self,Points,InterpolationNum):
        self.demension=Points.shape[1]   # 
        self.order=Points.shape[0]-1     # =-1
        self.num=InterpolationNum        # 
        self.pointsNum=Points.shape[0]   # 
        self.Points=Points
        
    # Bezeir
    def getBezierPoints(self,method):
        if method==0:
            return self.DigitalAlgo()
        if method==1:
            return self.DeCasteljauAlgo()
    
    # 
    def DigitalAlgo(self):
        PB=np.zeros((self.pointsNum,self.demension)) # 
        pis =[]                                      # 
        for u in np.arange(0,1+1/self.num,1/self.num):
            for i in range(0,self.pointsNum):
                PB[i]=(math.factorial(self.order)/(math.factorial(i)*math.factorial(self.order-i)))*(u**i)*(1-u)**(self.order-i)*self.Points[i]
            pi=sum(PB).tolist()                      #
            pis.append(pi)            
        return np.array(pis)
def normalization1(data):
    data = data - np.min(data, axis=0)
    data = data/(np.max(np.abs(data), axis=0)[np.newaxis, :]+0.00001)
    return data

with open('/home/jiashuncheng/code/MANN/plot/data/rsc_h_norm_witchout_b_2.pkl', 'rb') as a:
    x = pickle.load(a)
    target = pickle.load(a)
target = np.argmax(target, axis=1)
Y = np.stack(x).squeeze(2)[:,:,:]
Y = normalization1(Y)
P_data = Y[:, np.where(target==1)[0], :][:,:40,:]
N_data = Y[:, np.where(target==0)[0], :][:,:40,:]
N,T,S = 10,135,2
noise, n_samples = 0.2, 40
data = np.zeros((n_samples,N,S,T))
data[:,:,0,:] = P_data.transpose(1,2,0)
data[:,:,1,:] = N_data.transpose(1,2,0)
R = np.mean(data,0)
R -= np.mean(R.reshape((N,-1)),1)[:,None,None]
dpca = dPCA.dPCA(labels='st',n_components=3,regularizer='auto',join =[{'s': ['s', 'st']}, {'t'}])
dpca.protect = ['t']
print(R.shape, data.shape)
Z = dpca.fit_transform(R,data)
ZZ = Z['st'] # n_components, data[2], time

ZZ0 = Bezier(ZZ[:,0,:].T, 1800)
ZZ0 = ZZ0.getBezierPoints(0).T
ZZ1 = Bezier(ZZ[:,1,:].T, 1800)
ZZ1 = ZZ1.getBezierPoints(0).T
ratio = 1800/135

fig=plt.figure()
ax = fig.add_subplot(projection='3d')
y = ZZ0[:,:int(20*ratio)+1]
ax.scatter(y[0,:], y[1,:], y[2,:], marker='.', color='C2', label='Sample', zorder=2)
for i in range(y.shape[1]-1):
    ax.plot(y[0,i:i+2], y[1,i:i+2], y[2,i:i+2], color='C2', linewidth=2)
y = ZZ0[:,int(20*ratio):int(120*ratio)+1]
ax.scatter(y[0,:], y[1,:], y[2,:], marker='.', color='C0', label='Delay', zorder=1)
for i in range(y.shape[1]-1):
    ax.plot(y[0,i:i+2], y[1,i:i+2], y[2,i:i+2], color='C0', linewidth=2)
y = ZZ0[:,int(120*ratio):int(136*ratio)]
ax.scatter(y[0,:], y[1,:], y[2,:], marker='.', color='C3', label='Decision', zorder=2)
for i in range(y.shape[1]-1):
    ax.plot(y[0,i:i+2], y[1,i:i+2], y[2,i:i+2], color='C3', linewidth=2)

y = ZZ1[:,:int(20*ratio)+1]
ax.scatter(y[0,:], y[1,:], y[2,:], marker='.', color='C5', label='Sample', zorder=2)
for i in range(y.shape[1]-1):
    ax.plot(y[0,i:i+2], y[1,i:i+2], y[2,i:i+2], color='C5', linewidth=2)
y = ZZ1[:,int(20*ratio):int(120*ratio)+1]
ax.scatter(y[0,:], y[1,:], y[2,:], marker='.', color='C1', label='Delay', zorder=1)
for i in range(y.shape[1]-1):
    ax.plot(y[0,i:i+2], y[1,i:i+2], y[2,i:i+2], color='C1', linewidth=2)
y = ZZ1[:,int(120*ratio):int(136*ratio)]
ax.scatter(y[0,0], y[1,0], y[2,:], marker='.', color='C4', label='Decision', zorder=2)
for i in range(y.shape[1]-1):
    ax.plot(y[0,i:i+2], y[1,i:i+2], y[2,i:i+2], color='C4', linewidth=2)
plt.legend()
plt.show()
```
